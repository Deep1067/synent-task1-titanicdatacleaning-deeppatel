# Titanic Data Cleaning & Preprocessing

This notebook demonstrates a comprehensive data cleaning and preprocessing pipeline for the famous Titanic dataset. The goal is to transform the raw dataset into a clean, structured, and analysis-ready format, addressing common data quality issues such as missing values, incorrect data types, and duplicate records. This project is designed to be beginner-friendly, suitable for internship submissions, and ready for GitHub upload.

## 1. Dataset Overview

This section covers the necessary library imports and loading the `Titanic-Dataset.csv` file into a pandas DataFrame. We'll then perform a preliminary overview to understand its basic structure, including its shape, column names, and a quick look at the first few rows.

```python
import pandas as pd

file_path = 'Titanic-Dataset.csv'

try:
    df = pd.read_csv(file_path)
    print(f"Dataset '{file_path}' loaded successfully.")

    print("\n--- Performing Immediate Cleaning Steps ---")

    # 1. Handle Missing Values
    age_median_value = df['Age'].median()
    df['Age'] = df['Age'].fillna(age_median_value)
    print(f"  'Age' column missing values filled with median: {age_median_value:.2f}")

    embarked_mode_value = df['Embarked'].mode()[0]
    df['Embarked'] = df['Embarked'].fillna(embarked_mode_value)
    print(f"  'Embarked' column missing values filled with mode: '{embarked_mode_value}'")

    if 'Cabin' in df.columns:
        df = df.drop('Cabin', axis=1)
        print("  'Cabin' column dropped due to excessive missing values.")

    # 2. Duplicate Removal
    initial_rows = df.shape[0]
    duplicates_count = df.duplicated().sum()
    if duplicates_count > 0:
        df = df.drop_duplicates()
        print(f"  Successfully removed {duplicates_count} duplicate rows.")
    else:
        print("  No duplicate rows found to remove.")

    # 3. Data Type Conversion
    if 'Sex' in df.columns:
        df['Sex'] = df['Sex'].astype('category')
        print("  'Sex' column converted to 'category'.")
    if 'Embarked' in df.columns:
        df['Embarked'] = df['Embarked'].astype('category')
        print("  'Embarked' column converted to 'category'.")

    # 4. Renaming Columns
    df = df.rename(columns={
        'PassengerId': 'Passenger_ID',
        'Pclass': 'Passenger_Class'
    })
    print("  Columns 'PassengerId' and 'Pclass' renamed to 'Passenger_ID' and 'Passenger_Class'.")

    print("--- Immediate Cleaning Steps Completed ---")

except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found. Please ensure it's uploaded to the current directory.")
    df = None
```

```python
if df is not None:
    # Display the shape of the dataset (number of rows, number of columns)
    print("\n--- Dataset Shape ---")
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
if df is not None:
    # Display all column names to understand available features
    print("\n--- Column Names ---")
    print(df.columns.tolist())
```

```python
if df is not None:
    # Display the first 5 rows to get a quick glimpse of the data content and format
    print("\n--- First 5 Rows of the Dataset ---")
    display(df.head())
```

## 2. Dataset Information

Understanding the data types and non-null counts for each column is crucial for identifying potential data quality issues, such as missing values or incorrectly assigned data types. The `df.info()` method provides a concise summary of the DataFrame.

```python
if df is not None:
    print("--- Dataset Information (df.info()) ---")
    # The .info() method prints a summary of the DataFrame, including the index dtype and column dtypes,
    # non-null values, and memory usage.
    df.info()

    print("\nExplanation:")
    print("  - 'PassengerId', 'Survived', 'Pclass', 'SibSp', 'Parch', 'Fare' are integers or floats.")
    print("  - 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked' are object (string) types.")
    print("  - Notice the 'Age', 'Cabin', and 'Embarked' columns have fewer non-null entries than the total rows (891),")
    print("    indicating the presence of missing values that need to be addressed.")
```

## 3. Statistical Summary

A statistical summary provides key descriptive statistics for numerical columns, offering insights into the distribution, central tendency, and spread of the data. This helps in detecting outliers or unusual patterns.

```python
if df is not None:
    print("--- Statistical Summary (df.describe()) ---")
    # The .describe() method generates descriptive statistics that summarize the central tendency,
    # dispersion, and shape of a dataset's distribution, excluding NaN values.
    # This is applied to numerical columns by default.
    display(df.describe())

    print("\nExplanation of Statistical Summary:")
    print("  - `count`: Number of non-null observations. Confirms missing values in 'Age'.")
    print("  - `mean`, `std`: Average value and standard deviation. Useful for understanding data spread.")
    print("  - `min`, `max`: Minimum and maximum values. Can help identify extreme values or outliers.")
    print("  - `25%`, `50%` (median), `75%`: Quartiles. Provide insight into the distribution of data. For example, the median age is around 28-29.")
```

## 4. Missing Value Analysis

Identifying the exact count and percentage of missing values per column is the first step in deciding an appropriate strategy for handling them. This analysis highlights which columns require attention.

```python
if df is not None:
    print("--- Missing Values Analysis ---")
    # Calculate the number of missing values for each column
    missing_values_count = df.isnull().sum()
    # Calculate the percentage of missing values for each column
    missing_values_percentage = (df.isnull().sum() / len(df)) * 100

    # Create a DataFrame to display this information clearly
    missing_info_df = pd.DataFrame({
        'Missing Count': missing_values_count,
        'Missing Percentage (%)': missing_values_percentage
    })

    # Filter to show only columns with missing values and sort by count
    missing_info_df = missing_info_df[missing_info_df['Missing Count'] > 0].sort_values(by='Missing Count', ascending=False)

    if not missing_info_df.empty:
        print("Columns with Missing Values:")
        display(missing_info_df)
    else:
        print("No missing values found in the dataset.")
```

## 5. Handling Missing Values

Based on the analysis, we will implement strategies to handle missing values professionally:
-   **Age**: Impute with the median, as it's less sensitive to outliers than the mean.
-   **Embarked**: Impute with the mode, as it's a categorical feature.
-   **Cabin**: Drop this column entirely due to a very high percentage of missing values (approximately 77%), which makes imputation unreliable and the column potentially uninformative.

```python
if df is not None:
    # --- Before Handling Missing Values ---
    print("--- Missing Values Before Handling ---")
    missing_before_handling = df.isnull().sum()
    display(missing_before_handling[missing_before_handling > 0])

    # 1. Fill 'Age' with the median value
    # Median is robust to outliers, making it a good choice for numerical distributions that might be skewed.
    age_median_value = df['Age'].median()
    df['Age'] = df['Age'].fillna(age_median_value)
    print(f"\n'Age' column missing values filled with median: {age_median_value:.2f}")

    # 2. Fill 'Embarked' with the mode value
    # Mode is the most appropriate imputation strategy for categorical columns.
    embarked_mode_value = df['Embarked'].mode()[0] # .mode() returns a Series, so [0] selects the first mode
    df['Embarked'] = df['Embarked'].fillna(embarked_mode_value)
    print(f"'Embarked' column missing values filled with mode: '{embarked_mode_value}'")

    # 3. Drop 'Cabin' column due to excessive missing values
    # A large number of missing values often means the column lacks sufficient information.
    if 'Cabin' in df.columns:
        df = df.drop('Cabin', axis=1)
        print("'Cabin' column dropped due to excessive missing values.")
```

```python
if df is not None:
    # --- After Handling Missing Values ---
    print("\n--- Missing Values After Handling ---")
    missing_after_handling = df.isnull().sum()
    if missing_after_handling.sum() == 0:
        print("All specified missing values have been handled successfully. No missing values remain in the processed columns.")
    else:
        display(missing_after_handling[missing_after_handling > 0])
```

## 6. Duplicate Removal

Duplicate records can lead to biased statistical analysis and machine learning models. This step identifies and removes any exact duplicate rows to ensure the integrity of our dataset.

```python
if df is not None:
    # --- Check for duplicate records ---
    initial_rows_duplicates = df.shape[0]
    duplicates_count = df.duplicated().sum()
    print(f"--- Duplicate Records Before Removal ---")
    print(f"Number of duplicate rows found: {duplicates_count}\n")

    # --- Remove duplicates if present ---
    if duplicates_count > 0:
        df = df.drop_duplicates()
        print(f"Successfully removed {duplicates_count} duplicate rows.")
    else:
        print("No duplicate rows found to remove.")

    # --- Show before and after counts ---
    final_rows_duplicates = df.shape[0]
    print("\n--- Duplicate Records After Removal ---")
    print(f"Number of rows after removing duplicates: {final_rows_duplicates}")
    print(f"Number of duplicate rows remaining: {df.duplicated().sum()}")
    print(f"Total rows removed: {initial_rows_duplicates - final_rows_duplicates}")
```

## 7. Data Type Conversion

Optimizing data types can lead to significant memory savings and improved performance for certain operations. Categorical features like 'Sex' and 'Embarked' are best represented with the `category` data type.

```python
if df is not None:
    print("--- Data Types Before Conversion ---")
    display(df.dtypes)
```

```python
if df is not None:
    # Convert 'Sex' to 'category'
    # 'category' type is suitable for columns with a limited number of unique values, offering memory efficiency.
    if 'Sex' in df.columns:
        df['Sex'] = df['Sex'].astype('category')
        print("\n'Sex' column converted to 'category'.")

    # Convert 'Embarked' to 'category'
    if 'Embarked' in df.columns:
        df['Embarked'] = df['Embarked'].astype('category')
        print("'Embarked' column converted to 'category'.")
```

```python
if df is not None:
    print("\n--- Data Types After Conversion ---")
    display(df.dtypes)

    print("\nExplanation of changes:")
    print("  - 'Sex' and 'Embarked' have been converted from 'object' to 'category'.")
    print("  - This conversion is beneficial for memory optimization and can speed up operations on these columns.")
```

## 8. Renaming Columns

Renaming columns to more intuitive and consistent names enhances code readability and makes the dataset easier to work with, especially for new collaborators or future analysis.

```python
if df is not None:
    print("--- Column Names Before Renaming ---")
    print(df.columns.tolist())
```

```python
if df is not None:
    # Rename 'PassengerId' to 'Passenger_ID' and 'Pclass' to 'Passenger_Class'
    # Using a dictionary for renaming is a clear and concise way to handle multiple renames.
    df = df.rename(columns={
        'PassengerId': 'Passenger_ID',
        'Pclass': 'Passenger_Class'
    })
    print("\nColumns 'PassengerId' and 'Pclass' renamed to 'Passenger_ID' and 'Passenger_Class' respectively.")
```

```python
if df is not None:
    print("\n--- Column Names After Renaming ---")
    print(df.columns.tolist())
```
